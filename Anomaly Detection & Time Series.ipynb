# Anomaly Detection & Time Series

## Q1. What is Anomaly Detection? Explain its types with examples.

### **Anomaly Detection – Definition**

**Anomaly Detection** is the process of identifying data points, events, or observations that **deviate significantly from the normal pattern** of a dataset. These unusual patterns are called **anomalies**, **outliers**, or **exceptions**.

Anomaly detection is widely used in:
- Fraud detection in banking
- Network security (intrusion detection)
- Health monitoring systems
- Manufacturing defect detection
- System performance monitoring

Anomalies often indicate **critical incidents**, such as fraud, faults, errors, or rare events that require immediate attention.

---

## **Types of Anomalies**

Anomalies are generally classified into **three main types**:

1. **Point Anomaly**
2. **Contextual Anomaly**
3. **Collective Anomaly**

---

### **1. Point Anomaly (Global Outlier)**

#### **Definition:**
A single data point is considered anomalous if it is **far away from the rest of the data points**.

#### **Explanation:**
This is the simplest type of anomaly where one individual observation differs greatly from the majority of the data.

#### **Example:**
- In a dataset of employee salaries ranging from ₹25,000 to ₹80,000, if one entry shows ₹10,00,000, it is a point anomaly.
- A credit card transaction of ₹2,00,000 when the user usually spends ₹2,000–₹5,000.

#### **Real-world Use Case:**
- Fraud detection in banking transactions.
- Detecting abnormal sensor readings in IoT devices.

---

### **2. Contextual Anomaly (Conditional Outlier)**

#### **Definition:**
A data point is anomalous **in a specific context** but may be normal in another context.

#### **Explanation:**
The anomaly depends on contextual information such as **time, location, season, or user behavior**.

#### **Example:**
- Temperature of **35°C** is normal in summer but anomalous in winter.
- A person logging into a system at **3 AM** may be unusual, but normal for night-shift employees.

#### **Real-world Use Case:**
- Detecting unusual login times in cybersecurity.
- Identifying unusual electricity consumption at night.

---

### **3. Collective Anomaly**

#### **Definition:**
A collection of data points is anomalous **together**, even if individual points appear normal.

#### **Explanation:**
The anomaly is observed in a **sequence or group of data**, not in a single data point.

#### **Example:**
- A series of small transactions made rapidly from the same account may indicate fraud.
- A sequence of network packets that together indicate a cyber-attack pattern.

#### **Real-world Use Case:**
- Intrusion detection in networks.
- Detecting abnormal patterns in ECG signals in healthcare.

---

## **Comparison of Anomaly Types**

| Type                | What is Anomalous?               | Depends on Context? | Example |
|---------------------|-----------------------------------|----------------------|---------|
| Point Anomaly       | Single data point                | No                   | Salary ₹10,00,000 among ₹50,000 salaries |
| Contextual Anomaly  | Data point in a specific context | Yes                  | 35°C in winter |
| Collective Anomaly  | Group/sequence of points         | Sometimes            | Many small fraud transactions together |

---

## **Conclusion**

Anomaly detection plays a crucial role in identifying unusual patterns that may signal serious problems or opportunities. Understanding the types of anomalies—**point, contextual, and collective**—helps in selecting the right detection techniques for real-world applications such as fraud prevention, cybersecurity, healthcare monitoring, and industrial quality control.


## Q2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

Anomaly detection can be performed using different algorithms based on how they identify unusual data points. **Isolation Forest**, **DBSCAN**, and **Local Outlier Factor (LOF)** are three popular techniques used for detecting outliers, each with a different approach and ideal use case.

---

## **1. Isolation Forest**

### **Approach:**
- Based on the concept of **isolation** rather than distance or density.
- Randomly selects a feature and splits the data.
- Anomalies are isolated **faster** because they require fewer splits to separate from the rest of the data.
- Works well with high-dimensional datasets.

### **Key Idea:**
Outliers are rare and different, so they are easier to isolate.

### **Suitable Use Cases:**
- Fraud detection in banking transactions
- Network intrusion detection
- High-dimensional datasets
- Large datasets where speed is important

---

## **2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**

### **Approach:**
- Based on **density** of data points.
- Forms clusters where data points are densely packed.
- Points that do not belong to any cluster are considered **outliers (noise)**.

### **Key Idea:**
Normal points belong to dense regions; outliers lie in sparse regions.

### **Suitable Use Cases:**
- Geospatial data analysis
- Detecting noise in clustering problems
- Applications where clusters are irregular in shape
- Image processing and pattern recognition

---

## **3. Local Outlier Factor (LOF)**

### **Approach:**
- Based on **local density deviation** of a data point compared to its neighbors.
- Measures how isolated a point is with respect to surrounding data.
- Detects outliers by comparing local density.

### **Key Idea:**
An outlier has a significantly lower density than its neighbors.

### **Suitable Use Cases:**
- Detecting anomalies in datasets with varying density
- Credit card fraud detection
- Medical data analysis
- Situations where local context matters

---

## **Comparison Table**

| Feature / Aspect        | Isolation Forest                          | DBSCAN                                      | Local Outlier Factor (LOF)                    |
|--------------------------|--------------------------------------------|---------------------------------------------|-----------------------------------------------|
| **Basis**                | Isolation using random splits              | Density-based clustering                    | Local density comparison                      |
| **Main Concept**         | Outliers are easier to isolate             | Outliers lie in low-density regions         | Outliers have lower density than neighbors    |
| **Works with Clusters**  | No need for clusters                       | Yes, forms clusters                         | No clustering, compares neighbors             |
| **Handles High Dimensions** | Yes                                   | Not very efficient                          | Moderate                                      |
| **Computational Speed**  | Fast and scalable                         | Slower for large datasets                  | Moderate                                      |
| **Detects Local Outliers** | Limited                                 | Limited                                     | Very effective                                |
| **Output**               | Anomaly score                             | Noise points (outliers)                    | LOF score                                     |
| **Best Use Case**        | Large, high-dimensional data               | Spatial/geographical data                   | Data with varying densities                   |

---

## **Conclusion**

Isolation Forest, DBSCAN, and LOF use different strategies to detect anomalies. **Isolation Forest** is best for large and high-dimensional datasets, **DBSCAN** is ideal when clustering and noise detection are required, and **LOF** is most effective when detecting local anomalies in datasets with varying density. Choosing the right method depends on the nature of the dataset and the specific problem being solved.



## Q3. What are the key components of a Time Series? Explain each with one example.

A **Time Series** is a sequence of data points recorded at **regular time intervals** (hourly, daily, monthly, yearly, etc.). Time series analysis helps in understanding patterns over time and is widely used in finance, sales forecasting, weather prediction, and demand planning.

A time series is mainly composed of **four key components**:

1. **Trend**
2. **Seasonality**
3. **Cyclic Variation**
4. **Irregular (Random) Variation**

---

### **1. Trend**

#### **Definition:**
The **long-term movement** or direction in the data over a period of time. It shows whether the data is increasing, decreasing, or remaining constant.

#### **Example:**
- A company’s sales increasing year after year due to business growth.
- Population growth over decades.

**Insight:** Trend indicates the general direction of the data.

---

### **2. Seasonality**

#### **Definition:**
Regular and repeating patterns that occur at **fixed time intervals** due to seasonal factors such as weather, festivals, or holidays.

#### **Example:**
- Ice cream sales increasing every summer.
- Increased online shopping during Diwali or Christmas.

**Insight:** Seasonality repeats at known, fixed periods (daily, monthly, yearly).

---

### **3. Cyclic Variation**

#### **Definition:**
Fluctuations that occur over **long periods** but are **not of fixed frequency**. These are often influenced by economic or business cycles.

#### **Example:**
- Rise and fall of stock market over several years.
- Economic boom and recession phases.

**Insight:** Cycles are long-term and irregular in duration compared to seasonality.

---

### **4. Irregular (Random) Variation**

#### **Definition:**
Unpredictable, random variations caused by unexpected events.

#### **Example:**
- Sudden drop in sales due to a natural disaster.
- Market disruption due to a pandemic like COVID-19.

**Insight:** These variations cannot be predicted and do not follow a pattern.

---

## **Summary Table**

| Component                  | Description                                      | Example                                  |
|---------------------------|--------------------------------------------------|------------------------------------------|
| Trend                     | Long-term upward or downward movement            | Increasing yearly sales                  |
| Seasonality               | Repeating pattern at fixed intervals             | High sales during festivals              |
| Cyclic Variation          | Long-term fluctuations without fixed period      | Economic boom and recession              |
| Irregular Variation       | Random, unpredictable changes                    | Sales drop due to pandemic               |

---

## **Conclusion**

Understanding the key components of a time series—**Trend, Seasonality, Cyclic, and Irregular variations**—helps in accurate forecasting and better decision-making. By analyzing these components, businesses can predict future patterns and plan accordingly.


## Q4. Define Stationary in Time Series. How can you test and transform a non-stationary series into a stationary one?

### **What is Stationarity?**

A **time series is said to be stationary** if its statistical properties remain **constant over time**.

These properties include:
- **Mean** is constant
- **Variance** is constant
- **Autocorrelation** structure does not change over time

In simple words, the series does **not show trend or seasonality**, and its behavior is stable across time.

> Most time series forecasting models like **ARIMA** require the data to be stationary.

---

## **Characteristics of a Stationary Series**

- No upward or downward trend
- No seasonal patterns
- Constant fluctuations around a fixed mean
- Equal spread (variance) across time

---

## **How to Test Stationarity**

### **1. Visual Inspection (Graphical Method)**
- Plot the time series.
- Check for visible trend or seasonality.
- If the graph shows increasing/decreasing pattern → non-stationary.

### **2. Rolling Statistics**
- Calculate rolling mean and rolling variance.
- If these values change over time → non-stationary.

### **3. Augmented Dickey-Fuller (ADF) Test (Statistical Test)**

The **ADF test** is commonly used to check stationarity.

- **Null Hypothesis (H₀):** Series is non-stationary
- **Alternative Hypothesis (H₁):** Series is stationary

**Decision Rule:**
- If **p-value < 0.05** → Reject H₀ → Series is stationary
- If **p-value > 0.05** → Fail to reject H₀ → Series is non-stationary

---

## **How to Transform Non-Stationary Series into Stationary**

If a series is non-stationary, the following transformations can be applied:

### **1. Differencing**
- Subtract the previous observation from the current one.
- Removes trend.

\[
Y'_t = Y_t - Y_{t-1}
\]

**Example:** Used in ARIMA (Integrated part).

---

### **2. Log Transformation**
- Applied when variance increases over time.
- Stabilizes variance.

\[
Y'_t = \log(Y_t)
\]

---

### **3. Moving Average Smoothing**
- Removes noise and irregular fluctuations.

---

### **4. Seasonal Differencing**
- Subtract value from previous season.

\[
Y'_t = Y_t - Y_{t-s}
\]

Where **s** is seasonal period (e.g., 12 for monthly data).

---

### **5. Detrending**
- Remove the trend component using regression or decomposition.

---

## **Summary Table**

| Method                     | Purpose                          |
|---------------------------|----------------------------------|
| Differencing              | Remove trend                     |
| Log Transformation        | Stabilize variance               |
| Seasonal Differencing     | Remove seasonality               |
| Moving Average            | Smooth irregular variations      |
| Detrending                | Remove long-term movement        |
| ADF Test                  | Check stationarity statistically |

---

## **Conclusion**

Stationarity is a crucial requirement in time series analysis. By using visual methods, rolling statistics, and the ADF test, we can determine whether a series is stationary. If not, transformations like differencing, log transformation, and seasonal adjustments help convert the series into a stationary form suitable for forecasting models.


## Q5. Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.

Time series forecasting models such as **AR, MA, ARIMA, SARIMA, and SARIMAX** are widely used to model and predict future values based on past observations. These models differ in their structure, components, and use cases.

---

## **1. AR (AutoRegressive) Model**

### **Structure:**
- Uses past values of the time series to predict the current value.
- Denoted as **AR(p)** where *p* is the number of lag observations.

\[
Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \dots + \phi_p Y_{t-p} + \varepsilon_t
\]

### **Application:**
- When data is stationary and depends strongly on its past values.
- Example: Predicting stock prices based on previous prices.

---

## **2. MA (Moving Average) Model**

### **Structure:**
- Uses past error terms (residuals) to predict current value.
- Denoted as **MA(q)** where *q* is the number of lagged forecast errors.

\[
Y_t = c + \varepsilon_t + \theta_1 \varepsilon_{t-1} + \dots + \theta_q \varepsilon_{t-q}
\]

### **Application:**
- Useful when data has random shocks or noise.
- Example: Sensor data with random disturbances.

---

## **3. ARIMA (AutoRegressive Integrated Moving Average)**

### **Structure:**
- Combination of AR and MA with **differencing** to make data stationary.
- Denoted as **ARIMA(p, d, q)**:
  - *p* = AR order
  - *d* = differencing order
  - *q* = MA order

### **Application:**
- Non-stationary data that becomes stationary after differencing.
- Example: Sales forecasting with trend.

---

## **4. SARIMA (Seasonal ARIMA)**

### **Structure:**
- Extension of ARIMA that handles **seasonality**.
- Denoted as **SARIMA(p, d, q)(P, D, Q, s)**:
  - Seasonal AR, differencing, MA components
  - *s* = seasonal period (e.g., 12 for monthly data)

### **Application:**
- Data with repeating seasonal patterns.
- Example: Monthly electricity consumption, festival sales.

---

## **5. SARIMAX (Seasonal ARIMA with Exogenous Variables)**

### **Structure:**
- SARIMA model with **external (exogenous) variables** that influence the series.
- Includes additional predictors (X).

### **Application:**
- When external factors affect the time series.
- Example: Sales forecasting influenced by advertising spend, weather, or promotions.

---

## **Comparison Table**

| Model    | Full Form                              | Handles Trend | Handles Seasonality | Uses Past Values | Uses Past Errors | Uses External Variables | Suitable Application |
|----------|-----------------------------------------|---------------|---------------------|------------------|------------------|--------------------------|----------------------|
| AR       | AutoRegressive                         | No            | No                  | Yes              | No               | No                       | Stationary data with strong lag effect |
| MA       | Moving Average                         | No            | No                  | No               | Yes              | No                       | Data with noise/shocks |
| ARIMA    | AutoRegressive Integrated Moving Avg   | Yes           | No                  | Yes              | Yes              | No                       | Non-stationary data with trend |
| SARIMA   | Seasonal ARIMA                         | Yes           | Yes                 | Yes              | Yes              | No                       | Seasonal time series data |
| SARIMAX  | Seasonal ARIMA with Exogenous vars     | Yes           | Yes                 | Yes              | Yes              | Yes                      | Time series influenced by external factors |

---

## **Conclusion**

These models are selected based on the nature of the time series data. **AR** and **MA** are suitable for simple stationary data, **ARIMA** handles trend through differencing, **SARIMA** manages seasonality, and **SARIMAX** is most powerful when external variables impact the predictions.

 # Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series,and decompose it into trend, seasonality, and residual components





```python

# Create AirPassengers dataset locally (no internet), plot, and decompose

import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose

# Load AirPassengers dataset
df = pd.read_csv("AirPassengers.csv")

# Convert Month column to datetime and set as index
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)

series = df['#Passengers']

# Plot original series
plt.figure()
plt.plot(series)
plt.title('AirPassengers Time Series')
plt.xlabel('Time')
plt.ylabel('Number of Passengers')
plt.show()

# Decomposition
decomposition = seasonal_decompose(series, model='multiplicative', period=12)

# Trend
plt.figure()
plt.plot(decomposition.trend)
plt.title('Trend Component')
plt.show()

# Seasonality
plt.figure()
plt.plot(decomposition.seasonal)
plt.title('Seasonal Component')
plt.show()

# Residuals
plt.figure()
plt.plot(decomposition.resid)
plt.title('Residual Component')
plt.show()
```

# Output Explanation

**Original Series**: Shows increasing number of passengers over time with visible seasonal peaks.

**Trend Component**: Displays the long-term upward movement in passenger count.

**Seasonal Component**: Shows repeating yearly pattern (seasonality).

**Residual Component**: Represents irregular/random fluctuations after removing trend and seasonality.




# Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) todetect anomalies. Visualize the anomalies on a 2D scatter plot.


```python

# Apply Isolation Forest on a synthetic NYC Taxi Fare-like dataset and visualize anomalies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Create synthetic taxi fare dataset
np.random.seed(42)
n = 500

distance = np.random.normal(5, 2, n).clip(0.5, 15)  # trip distance in km
fare = distance * 50 + np.random.normal(0, 20, n)   # fare roughly proportional to distance

# Inject anomalies
distance_anom = np.random.uniform(1, 15, 20)
fare_anom = np.random.uniform(1000, 3000, 20)

distance = np.concatenate([distance, distance_anom])
fare = np.concatenate([fare, fare_anom])

data = pd.DataFrame({
    "distance_km": distance,
    "fare_amount": fare
})

# Isolation Forest
iso = IsolationForest(contamination=0.05, random_state=42)
data["anomaly"] = iso.fit_predict(data[["distance_km", "fare_amount"]])

# Separate normal and anomalies
normal = data[data["anomaly"] == 1]
anomalies = data[data["anomaly"] == -1]

# Scatter plot
plt.figure()
plt.scatter(normal["distance_km"], normal["fare_amount"])
plt.scatter(anomalies["distance_km"], anomalies["fare_amount"])
plt.xlabel("Distance (km)")
plt.ylabel("Fare Amount")
plt.title("Isolation Forest Anomaly Detection on Taxi Fare Data")
plt.show()

len(anomalies)
```

# Output

The scatter plot shows most data points forming a linear pattern (normal fares).

Points far away from this pattern are detected as **anomalies** by Isolation Forest.

Total anomalies detected: **26**

# Question 8: Train a SARIMA model on the monthly airline passengers dataset.Forecast the next 12 months and visualize the results.



```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load AirPassengers dataset from file
df = pd.read_csv("AirPassengers.csv")
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)

# Train SARIMA model
model = SARIMAX(df['#Passengers'],
                order=(1,1,1),
                seasonal_order=(1,1,1,12))
results = model.fit(disp=False)

# Forecast next 12 months
forecast = results.forecast(12)

# Plot
plt.figure()
plt.plot(df['#Passengers'], label="Observed")
plt.plot(forecast, label="Forecast (12 months)")
plt.legend()
plt.title("SARIMA on AirPassengers Dataset")
plt.show()
```

**Conclusion**

The SARIMA model effectively captures both trend and seasonal patterns in the AirPassengers dataset and provides a reasonable forecast for the next year.

# Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Step 1: Create a numerical dataset
np.random.seed(42)
X_normal = np.random.normal(loc=50, scale=5, size=(100, 2))
X_outliers = np.random.normal(loc=80, scale=1, size=(5, 2))
X = np.vstack([X_normal, X_outliers])

# Step 2: Apply Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(X)

# Step 3: Separate normal points and outliers
normal = X[y_pred == 1]
outliers = X[y_pred == -1]

# Step 4: Visualization
plt.figure()
plt.scatter(normal[:, 0], normal[:, 1])
plt.scatter(outliers[:, 0], outliers[:, 1])
plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Output
print("Total data points:", len(X))
print("Detected outliers:", len(outliers))
```

**Conclusion**

LOF successfully detected anomalous points by comparing local density differences, and the visualization clearly highlights the outliers.
